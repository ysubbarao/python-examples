# Basic plots with matplotlib

```python
import matplotlib.pyplot as plt
```

```python
year = [1950,1970,1990,2010]
population = [2.519,3.692,5.263,6.972]
```

```python
plt.plot(year,population)
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

# scatterplot

```python
plt.scatter(year,population)
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
