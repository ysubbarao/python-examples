```python
#Multiple functions can drawn in the same line
```

```python
x = range(10)
plt.plot(x, [i for i in x],x,[i*i for i in x],x,[i*i*i for i in x])
plt.show()
```

```python
#Adding a grid

x= range(10)
plt.plot(x,[i for i in x],
         x,[i*i for i in x],
         x,[i*i*i for i in x])
print(x)
plt.grid(True)
plt.axis([-1,5,-1,10])   #Limiting the axis
plt.show()
```

```python
# Limit the axis
```

```python
x = range(10)
plt.plot(x,[i for i in x],
         x,[i*i for i in x],
         x,[i*i*i for i in x])
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.show()
```

```python
# Adding labels and title
```

```python
x = range(10)
plt.plot(x,[i for i in x],
         x,[i*i for i in x],
         x,[i*i*i for i in x])
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-2,10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
plt.show()
```

```python
#Adding legends (adding lable in each line on the graph)
```

```python
x = range(5)
plt.plot(x,[i for i in x],label = 'linear')
plt.plot(x,[i*i for i in x],label = 'square')
plt.plot(x,[i*i*i for i in x],label='cube')
         
         
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend()
plt.show()
```

```python
#Barchart
```

```python
plt.bar([1,2,3],[1,2,3])
plt.show()
```

```python
dict_ob = {'A':10,'B':20,'C':30,'D':40}
for i,key in enumerate(dict_ob):
    plt.bar(i,dict_ob[key])
    print('i value is :',i)
    print('Key value is: ',dict_ob[key])
    plt.show()
```

```python
#dict_ob = {'A':10,'B':20,'C':30,'D':40}
for i,key in enumerate(dict_ob):
    plt.bar(i,dict_ob[key])
    #print('i value is :',i)
    #print('Key value is: ',dict_ob[key])
    plt.xticks(np.arange(len(dict_ob)),dict_ob.keys())
    plt.show()
```

```python
#Pie charts are used to compare multiple parts against the whole
```
