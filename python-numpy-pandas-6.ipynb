```python
import pandas as pd
```

```python
student_details={'sid':[1,2,3,4,5],
                  'sname':['A','B','C','D','E'],
                'region':['N','S','E','E','W']}

student_marks ={'sid':[1,2,3,4,6],
                'sub':['Py','R','Py','Tab','Py'],
                'marks':[43,23,44,22,32]
               }
```

```python
student_details
```

```python
student_marks
```

```python
df1=pd.DataFrame(student_details)
df1
```

```python
df2=pd.DataFrame(student_marks)
df2
```

```python
#Join two tables
#Inner join -- common records (1,2,3,4) from above df1 and df2,remaining records left.
pd.merge(df1,df2,on='sid')
```

```python
#Left outer join --all the records form the left table(df1:1,2,3,4,5) matched with right table df2(1,2,3,4)
pd.merge(df1,df2,on ='sid',how='left')

```

```python
#Right outer join--all the records form the right table(df1:1,2,3,4,6) matched with left table df1(1,2,3,4)
pd.merge(df1,df2,on='sid',how ='right')
```

```python
#Full outer join--all the records from both table can participate df1,df2
pd.merge(df1,df2,on='sid',how='outer')
```

```python
student_details={'s_id':[1,2,3,4,5],
                  'sname':['A','B','C','D','E'],
                'region':['N','S','E','E','W']}

student_marks ={'sid':[1,2,3,4,6],
                'sub':['Py','R','Py','Tab','Py'],
                'marks':[43,23,44,22,32]
               }
```

```python
df1 = pd.DataFrame(student_details)
df2 = pd.DataFrame(student_marks)
df1
```

```python
df2
```

```python
#inner join
df1.merge( df2, left_on='s_id', right_on='sid')
```

```python
#full outer join
df1.merge(df2,left_on='s_id',right_on='sid', how='outer')
```

```python
pd.merge(df1,df2,left_on='s_id',right_on='sid',how='outer')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.merge(pd.merge(df1,df2,left_on='s_id',right_on='sid'),df3,left_on='s_id',right_on='sid')
```
